[ss.ge](https://home.ss.ge/ka/udzravi-qoneba) scraper

```python
properties = []
total_pages = 5
for page in range(1, total_pages + 1):
    print(f"Scrapign page: {page} - {total_pages}...")
    response = requests.get(f"https://home.ss.ge/ka/udzravi-qoneba/l/bina/iyideba?cityIdList=95&currencyId=1&page={page}")
    soup = BeautifulSoup(response.text, "html.parser")

    tags = soup.select("div.sc-1384a2b8-3.bZedIp.top-grid-lard > div.jlmink > a")

    for tag in tags:
        title = tag.select_one("h2.listing-detailed-item-title").text
        price_tag = tag.select_one("span.listing-detailed-item-price")
        if price_tag is None:
            price = ""
        else:
            price = price_tag.text
        properties.append({
            "title": title,
            "price": price
        })
```

```python
# response.cookies["ss-session-token"]
response = requests.post(
    "https://api-gateway.ss.ge/v1/RealEstate/LegendSearch",
    headers={
        "Authorization": f"BEARER {token}"
    },
    json={
        "realEstateType":5,
        "realEstateDealType":4,
        "cityIdList":[95],
        "currencyId":1,
        "page":2,
        "pageSize":16
    }
)
```

```python
raw_properties = response.json()["realStateItemModel"]
properties = []
for prop in raw_properties:
    properties.append({
        "district": prop["address"]["districtTitle"],
        "price": prop["price"]["priceGeo"],
        "bedrooms": prop["numberOfBedrooms"],
        "area": prop["totalArea"],
        "total_floors_in_building": prop["totalAmountOfFloor"],
        "apartment_floor": prop["floorNumber"]
    })
```
